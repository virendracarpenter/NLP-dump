```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('us_births_2016_2021.csv')
df = df.drop(['State Abbreviation','Education Level of Mother','Average Age of Mother (years)'], axis=1)
df.head(20)

```

```python
def year(year):
    if year < int(df.iloc[1]['Year']) or int(year > df.iloc[-1]['Year']):
        print('Invalid Input !!!')
        return None
        
    df_year = df[df['Year'] == int(year)]
    gender_count = df_year.groupby(['Gender']).size()
    stats = df_year['Number of Births'].describe()
    
    
         
    print(gender_count)
    print(stats)

def list_data():
    print(df)
    
def main():
    while True:
        i = int(input("""Choose Number from the below option\n
                            1. To List All Data.\n
                            2. To Find Specific Year Details.\n
                            3. To Find The Most Populated Country by Year.\n
                            4. To Find The Top Education Level Code Countries.\n
                            5. To Find The Statatics by Country Name.\n
                            0. To exit.\n
                    """))
        if i == 1:
            list_data()
        elif i == 2:
            y = input("Enter a year: ")
            year(y)
        elif i == 3:
            pop_country = df.groupby(['Year', 'State'])['Number of Births'].sum().groupby('Year').idxmax()
            print(pop_country)
        elif i == 4:
            n = int(input("Enter a number: "))
            top_edu = df.groupby(['State', 'Education Level Code'])['Number of Births'].sum().groupby('State').nlargest(n)
            print(top_edu)
        elif i == 5:
            c = input("Enter a country name: ")
            stats = df[df['State'] == c][['Number of Births', 'Average Birth Weight (g)']].describe()
            print(stats)
        else:
            break


main()
```
