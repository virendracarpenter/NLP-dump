```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('us_births_2016_2021.csv')
df = df.drop(['State Abbreviation','Education Level of Mother','Average Age of Mother (years)'], axis=1)
df.head(20)
```

```python
def year(year):
    df_year = df[df['Year'] == int(year)]
    gender_count = df_year.groupby(['Gender']).size()
    stats = df_year[['Education Level Code', 'Number of Births', 'Average Birth Weight (g)']].describe()
     
    
    print(gender_count)
    print(stats)
year(2016)
```
