```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../ML/homeprice.csv')

def predict_price(area):

    x = df["area"]
    y = df["price"]

    coef = np.polyfit(x, y, 1)
    print('coef' ,coef)
        
    pp = coef[0] * area + coef[1]
    # y = mx + b
    return pp


print(predict_price(1000))
print(predict_price(1500))
print(predict_price(2000))
x = [1000,1500,2000]
y = [predict_price(x[0]), predict_price(x[1]),predict_price(x[2])]

plt.scatter(x,y)
plt.show()

```
