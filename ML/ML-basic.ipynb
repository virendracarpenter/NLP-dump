```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprice.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price, color = 'red', marker = 'x')
```

```python
# y = mx + b
# price = m * area + b

reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price, color = 'red', marker = 'x')
plt.plot(df.area,reg.predict(df[['area']]), color = 'blue')
```

```python
d = pd.read_csv('areas.csv')
d.head()
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
d.to_csv('predictprices.csv', index=False)
```

```python
df = pd.read_csv('canada_pci.csv')
df.head()
```

```python
plt.plot(df['year'],df['pci'], 'x')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.pci)

```

```python
plt.scatter(df.year, df.pci, color = 'red', marker = 'x')
plt.plot(df.year,reg.predict(df[['year']]), color = 'blue')
plt.show()






# plt.scatter(df.area,df.price, color = 'red', marker = 'x')
# plt.plot(df.area,reg.predict(df[['area']]), color = 'blue')
```
